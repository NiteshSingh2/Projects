## Let's Get started With spacy

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords= list(STOP_WORDS)
stopwords
```

Let's go ahead and create the NLP model

```python
nlp= spacy.load('en_core_web_sm')
```

```python
doc= nlp(text)
```

```python
tokens= [token.text for token in doc]   #One think i notice here is that Punctuation and stopwords are also part of original tokens.
print(tokens)
```

## Removed Punctuation and stopwords

```python
punctuation
```

We need to calculate each of these words how many times has occurred in this text and them We based on those frequency.We will identify that the most important sentence.

```python
word_freq= {}
for word in doc:
    if word.text.lower() not in stopwords:
      if word.text.lower() not in punctuation:
        if word.text not in word_freq.keys():
          word_freq[word.text]=1
        else:
          word_freq[word.text] += 1
```

```python
word_freq
```

```python
max_freq= max(word_freq.values())   # so here we have max freq that is 34
max_freq
```

We attempt to reduce its randomness, bringing it closer to a predefined “standard”. This helps us to reduce the amount of different information that the computer has to deal with, and therefore improves efficiency.


```python
for word in word_freq.keys():                          #normalized freq can be achieved 
  word_freq[word] = word_freq[word]/max_freq
```

```python
word_freq  
```

```python
sentence_tokens = [sent for sent in doc.sents]
sentence_tokens
```

So, we are going to calculate the sentence score

```python
sentence_scores={}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_freq.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_freq[word.text.lower()]
            else:
              sentence_scores[sent] += word_freq[word.text.lower()] 
```

So,here we have got the sentence score for each of these sentence.

```python
sentence_scores
```

The heapq library has been used to pick the top 7 sentences to summarize the article.

```python
from heapq import nlargest
```

```python
select_len= int(len(sentence_tokens)*0.3) #it will tell us total length of sentences present.
select_len  
```

so the 30% of total sentence is 6 

```python
summary = nlargest(select_len, sentence_scores, key = sentence_scores.get)
```

```python
summary
```

This is the summary of text

Extraction-based Summarization works by extracting key phrases in the given text and joining them to form meaningful sentences whereas Abstraction-based Summarization works by creating new phrases and sentences that resonate with the meaning of the original text. Both are supervised machine learning problems.

## Create Tokenizer

```python
# Load tokenizer
tokenizer = PegasusTokenizer.from_pretrained("google/pegasus-xsum")
```

```python
# Load model 
model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
```

# Perform Abstractive Summarization

```python
Text="""With more than a decade of use and 20 billion EV charging miles to its name, the Tesla charging connector is the most proven in North America, offering AC charging and up to 1 MW DC charging in one slim package. It has no moving parts, is half the size, and twice as powerful as Combined Charging System (CCS) connectors.

In pursuit of our mission to accelerate the world’s transition to sustainable energy, today we are opening our EV connector design to the world. We invite charging network operators and vehicle manufacturers to put the Tesla charging connector and charge port, now called the North American Charging Standard (NACS), 
on their equipment and vehicles. NACS is the most common charging standard in North America:
NACS vehicles outnumber CCS two-to-one, and Tesla's Supercharging network has 60% more NACS posts than all the CCS-equipped networks combined."""
```

So,first we can do is convert this text into token representation.

```python
# Create tokens - number representation of our text                                     
tokens = tokenizer(Text, truncation=True, padding="longest", return_tensors="pt")
```

Truncation-This is basically means that it's going to shorten our text to make sure that it's of an appropriate length.

Padding-padding to be as long as possible so it was set padding equal to longest

Return_tensors = “pt” is just for the tokenizer to return PyTorch tensors.

So as you can see the tokens representation of our body of text so basically what we're saying is that all of this here is represented as these numbers to be able to pass  it through to our pegasus model.

# Summarize

```python
summary = model.generate(**tokens)
```
